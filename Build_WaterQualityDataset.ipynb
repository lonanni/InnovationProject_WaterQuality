```python
import numpy as np
import sys
import pandas as pd
```

# Download files https://environment.data.gov.uk/water-quality/view/download/new

```python
dir_files = "./data/"
```

```python
dfs = {}

# Load the CSV into the dictionary 

for year in (range(2001, 2025)):
    file_path = f'{dir_files}/SSD-{year}.csv'  
    dfs[year] = pd.read_csv(file_path, engine='python')
```

```python
# Concatenate the datasets
df_to_conc = []

for year in (range(2001, 2025)):
    file_path = f'{dir_files}/SSD-{year}.csv'  
    df_to_conc.append(pd.read_csv(file_path, engine='python'))
```

```python
df_water = pd.concat(df_to_conc, ignore_index=True) 

```

```python
df_water.to_csv(dir_files+"WaterQuality.csv")
```
